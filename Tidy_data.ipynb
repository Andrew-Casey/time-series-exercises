# 1
Attendance

-Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
df = pd.read_clipboard()
df
```

```python
df = df.reset_index()
df
```

```python
df = df.rename(columns= {'index':'name'})
df
```

```python
df_a = pd.melt(df, id_vars='name', var_name='day', value_name='attendance')
df_a.head()
```

```python
# Convert attendance values to numerical values
attendance_values = {'P': 1, 'T': 0.5, 'A': 0}
df_a['attendance'] = df_a['attendance'].map(attendance_values)

# Group by student name and calculate attendance percentage
attendance_percentage = df_a.groupby('name')['attendance'].mean()

# Display the attendance percentages
print(attendance_percentage)
```

# 3 
Cake Recipes

- Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
- Tidy the data as necessary.
- Which recipe, on average, is the best?
- Which oven temperature, on average, produces the best results?
- Which combination of recipe, rack position, and temperature gives the best result?

```python
df_cr = pd.read_clipboard()
df_cr.head()
```

```python
# Tidy the data by separating the 'recipe:position' column
df_cr[['recipe', 'position']] = df_cr['recipe:position'].str.split(':', expand=True)
df_cr
```

```python
# Drop the 'recipe:position' column
df_cr.drop('recipe:position', axis=1, inplace=True)
df_cr
```

```python
# Tidy the data by melting the dataframe
tidy_df = pd.melt(df_cr, id_vars=['recipe', 'position'], var_name='temperature', value_name='score')
tidy_df.head()
```

```python
# Calculate the average score for each recipe
avg_recipe_scores = tidy_df.groupby('recipe')['score'].mean()
avg_recipe_scores
```

```python
# Determine the best recipe
best_recipe = avg_recipe_scores.idxmax()
best_recipe
```

```python
# Calculate the average score for each oven temperature
avg_temp_scores = tidy_df.groupby('temperature')['score'].mean()
avg_temp_scores
```

```python
# Determine the best oven temperature
best_temperature = avg_temp_scores.idxmax()
best_temperature
```

```python
# Determine the combination that gives the best result
best_combination = tidy_df[tidy_df['score'] == tidy_df['score'].max()].iloc[0]
best_combination
```

```python
# Display the results
print("Average scores for each recipe:")
print(avg_recipe_scores)
print("\nThe best recipe is:", best_recipe)

print("\nAverage scores for each oven temperature:")
print(avg_temp_scores)
print("\nThe best oven temperature is:", best_temperature)

print("\nThe combination that gives the best result is:")
print(best_combination)
```
